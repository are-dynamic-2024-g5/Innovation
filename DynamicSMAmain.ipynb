```python
import numpy as np
import matplotlib.pyplot as plt

class Agent:
    def __init__(self, adoption_rate, age, social_influence, communication_strategy, competition):
        self.adoption_rate = adoption_rate
        self.age = age
        #self.simplicity = simplicity
        self.social_influence = social_influence
        self.communication_strategy = communication_strategy
        #self.context = context
        self.competition = competition
        self.adopted = False

def simulate_diffusion(num_agents, initial_adoption_rate, num_steps, age_threshold):
    # Initialize agents
    agents = [Agent(initial_adoption_rate, np.random.rand()*100, 15, np.random.rand(), np.random.rand()) for _ in range(num_agents)]

    # Simulation
    adoption_counts = []
    adoption_speed = []
    adoption_age_range = []
    for _ in range(num_steps):
        adopted_agents = sum(agent.adopted for agent in agents)
        adoption_counts.append(adopted_agents)
        adoption_speed.append(adopted_agents / num_agents)  # Adoption speed

        # Calculate number of adoptions in specified age range
        adoption_in_age_range = sum(1 for agent in agents if not agent.adopted and agent.age <= age_threshold)
        adoption_age_range.append(adoption_in_age_range)

        for agent in agents:
            if not agent.adopted:
                # Calculate adoption probability based on parameters
                adoption_prob = (
                    agent.adoption_rate  * agent.social_influence *
                    agent.communication_strategy * agent.context  (1 - agent.competition)
                )
                # Adoption decision
                if np.random.rand() < adoption_prob:
                    agent.adopted = True

    not_adopted_count = num_agents - adoption_counts[-1]  # Number of people who haven't adopted
    return adoption_counts, not_adopted_count, adoption_speed, adoption_age_range

# Parameters
num_agents = 10000
initial_adoption_rate = 0.1
num_steps = 50
age_threshold = 50  # Choose the age threshold for the specified age range

# Simulation
adoption_counts, not_adopted_count, adoption_speed, adoption_age_range = simulate_diffusion(num_agents, initial_adoption_rate, num_steps, age_threshold)

# Plot results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(range(num_steps), adoption_counts)
plt.xlabel('Time Steps')
plt.ylabel('Number of Adopted Agents')
plt.title('Diffusion of Innovation')

plt.subplot(1, 2, 2)
plt.plot(range(num_steps), adoption_speed)
plt.xlabel('Time Steps')
plt.ylabel('Adoption Speed')
plt.title('Adoption Speed Over Time')

plt.tight_layout()
plt.show()

print("Final number of adopted agents:", adoption_counts[-1])
print("Number of agents who haven't adopted:", not_adopted_count)

# Display number of adoptions in specified age range
print("Number of adoptions in age range (<= {}):".format(age_threshold), adoption_age_range[-1])
```
