```python
import numpy as np
import matplotlib.pyplot as plt

class Agent:
    def __init__(self, x, y, threshold):
        self.x = x
        self.y = y
        self.threshold = threshold
        self.innovator = False

    def distance_to(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def is_close_to(self, other):
        return self.distance_to(other) <= self.threshold

def main():
    # Parameters
    num_agents = 100
    grid_size = 100
    threshold = 10
    innovation_rate = 0.1
    num_steps = 100

    # Initialize agents
    agents = [Agent(np.random.uniform(0, grid_size), np.random.uniform(0, grid_size), threshold) for _ in range(num_agents)]

    # Introduce innovators
    num_innovators = int(num_agents * innovation_rate)
    for i in range(num_innovators):
        agents[i].innovator = True

    # Simulation
    innovators = []
    adopters = []
    for _ in range(num_steps):
        innovators.append(sum(agent.innovator for agent in agents))
        adopters.append(len(agents) - innovators[-1])
        for agent in agents:
            if agent.innovator:
                for other in agents:
                    if not other.innovator and agent.is_close_to(other):
                        other.innovator = True

    # Plot results
    plt.plot(range(num_steps), innovators, label='Innovators')
    plt.plot(range(num_steps), adopters, label='Adopters')
    plt.xlabel('Time Steps')
    plt.ylabel('Number of Agents')
    plt.title('Diffusion of Innovation')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Agent:
    def __init__(self, x, y, threshold, adoption_rate, social_network_influence):
        self.x = x
        self.y = y
        self.threshold = threshold
        self.adoption_rate = adoption_rate
        self.social_network_influence = social_network_influence
        self.adopted = False

    def distance_to(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

    def is_close_to(self, other):
        return self.distance_to(other) <= self.threshold

def main():
    # Parameters
    num_agents = 100
    grid_size = 100
    threshold = 10
    innovation_rate = 0.1
    num_steps = 100
    adoption_rate = 0.1
    social_network_influence = 0.1

    # Initialize agents
    agents = [Agent(np.random.uniform(0, grid_size), np.random.uniform(0, grid_size), threshold, adoption_rate, social_network_influence) for _ in range(num_agents)]

    # Introduce innovators
    num_innovators = int(num_agents * innovation_rate)
    for i in range(num_innovators):
        agents[i].adopted = True

    # Simulation
    adopted_counts = []
    for _ in range(num_steps):
        adopted_counts.append(sum(agent.adopted for agent in agents))
        for agent in agents:
            if not agent.adopted:
                # Influence from social network
                neighbors = [other for other in agents if agent.is_close_to(other)]
                neighbors_influence = sum(other.adopted for other in neighbors) / len(neighbors) if neighbors else 0
                # Adoption probability
                adoption_prob = agent.adoption_rate + agent.social_network_influence * neighbors_influence
                # Adoption decision
                if np.random.rand() < adoption_prob:
                    agent.adopted = True

    # Plot results
    plt.plot(range(num_steps), adopted_counts, label='Adopted Count')
    plt.xlabel('Time Steps')
    plt.ylabel('Number of Adopted Agents')
    plt.title('Diffusion of Innovation')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class Agent:
    def __init__(self, adoption_rate, simplicity, social_influence, communication_strategy, context, competition):
        self.adoption_rate = adoption_rate
        self.simplicity = simplicity
        self.social_influence = social_influence
        self.communication_strategy = communication_strategy
        self.context = context
        self.competition = competition
        self.adopted = False

def simulate_diffusion(num_agents, initial_adoption_rate, num_steps):
    # Initialize agents
    agents = [Agent(initial_adoption_rate, np.random.rand(), np.random.rand(), np.random.rand(), np.random.rand(), np.random.rand()) for _ in range(num_agents)]

    # Simulation
    adoption_counts = []
    for _ in range(num_steps):
        adoption_counts.append(sum(agent.adopted for agent in agents))
        for agent in agents:
            if not agent.adopted:
                # Calculate adoption probability based on parameters
                adoption_prob = (
                    agent.adoption_rate * agent.simplicity * agent.social_influence *
                    agent.communication_strategy * agent.context * (1 - agent.competition)
                )
                # Adoption decision
                if np.random.rand() < adoption_prob:
                    agent.adopted = True

    return adoption_counts

# Parameters
num_agents = 1000
initial_adoption_rate = 0.1
num_steps = 50

# Simulation
adoption_counts = simulate_diffusion(num_agents, initial_adoption_rate, num_steps)

# Plot results
plt.plot(range(num_steps), adoption_counts)
plt.xlabel('Time Steps')
plt.ylabel('Number of Adopted Agents')
plt.title('Diffusion of Innovation')
plt.grid(True)
plt.show()
```
